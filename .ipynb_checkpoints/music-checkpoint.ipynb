Из тех данных, которые у нас есть, я выделил информацию, которую хочу получить:

А. Характеристики треков в стиле техно:
    1. качество песни (соотношение среднего времени прослушивания трека к его длительности)
    2. популярность песни (количество прослушиваний за последний месяц) + составить топ-30
    (посмотреть корреляцию между популярностью и дослушиванием, просто интересно)

Б. Характеристики пользователей:
    1. как много слушает техно (соотношение количества прослушанных техно-треков к общему количеству прослушанных треков)
    2. насколько популярные песни слушает (количество треков из топ-30 в прослушанных за последний месяц треках)
    3. Насколько долго слушает треки? (соотношение среднего времени прослушивания трека пользователем к общему среднему времени прослушивания трека (А.1))
    
На основе этих характеристик потом можно составить score для каждого из пользователей. На основе этого score будет формироваться выборка для отправки email.

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (15, 5)

# Подготовка данных
data_all = pd.read_csv("music_data.csv")
data_all.set_index('Unnamed: 0', drop=True, inplace=True)
data_all.index.name = 'id'

# Берем только свежие данные 
# (за последние 30 дней относительно момента окончания последнего прослушивания)
data = data_all[pd.to_datetime(data_all['utc_audition_end_dttm']) > (pd.to_datetime(data_all['utc_audition_end_dttm']).max() - pd.to_timedelta(30, unit='d'))].copy()
techno_data = data[data['track_genre']=='TECHNO'].copy()

data.columns
```

```python
# Процент времени прослушивания от длины трека - techno_data['l_time_to_t_duration']
# Средний процент прослушивания треков (качество) -  track_quality['mean_l_time_to_duration']
```

```python
techno_data['l_time_to_t_duration'] = (pd.to_datetime(techno_data['utc_audition_end_dttm']) - pd.to_datetime(techno_data['utc_audition_start_dttm'])) / pd.to_timedelta(techno_data['track_duration'], unit='s') * 100.0
track_quality = techno_data[['track_id', 'l_time_to_t_duration']].groupby('track_id').mean().copy()
track_quality.columns = ['mean_l_time_to_duration']

track_quality.sort_values(by='mean_l_time_to_duration', ascending=False).reset_index(drop=True).plot()
```

```python
# Количество прослушиваний техно-треков (популярность) - track_popularity['auditions_count']
```

```python
track_popularity = techno_data[['track_id', 'audition_id']].groupby('track_id').count().copy()
track_popularity.columns = ['auditions_count']

track_popularity.sort_values(by='auditions_count', ascending=False).reset_index(drop=True).plot()
```

```python
# Ради интереса решил визуализировать корреляцию между качеством и популярностью - ее нет =)
view = track_quality.join(track_popularity, how='inner', on='track_id')
view.sort_values(by='mean_l_time_to_duration', ascending=False).plot()
```

```python
# Процент прослушиваний техно к общему количеству прослушиваний для пользователей
# - user_techno_to_all_tracks['techno_to_all_percent']
```

```python
user_techno_tracks = techno_data[['user_id', 'audition_id']].groupby('user_id').count()
user_techno_tracks.columns = ['techno_count']
user_all_tracks = data[['user_id', 'audition_id']].groupby('user_id').count()
user_all_tracks.columns = ['all_count']

user_techno_to_all_tracks = user_techno_tracks.join(user_all_tracks, how='inner', on='user_id')
user_techno_to_all_tracks['techno_to_all_percent'] = user_techno_to_all_tracks['techno_count'] / user_techno_to_all_tracks['all_count'] * 100.0

user_techno_to_all_tracks.sort_values(by='techno_to_all_percent', ascending=False)['techno_to_all_percent'].reset_index(drop=True).plot()
```

```python
# Процент от топ-30 популярных техно-треков среди всех прослушанных пользователем техно-треков
# - user_thirty_popular_percentage['thirty_pop_percent']
```

```python
top_thirty = track_popularity.sort_values(by='auditions_count', ascending=False)[:30]
user_thirty_popular_percentage = techno_data.join(top_thirty, how='inner', on='track_id')[['user_id', 'track_id']].groupby('user_id').count()
user_thirty_popular_percentage.columns = ['thirty_pop_percent']
user_thirty_popular_percentage['thirty_pop_percent'] = user_thirty_popular_percentage['thirty_pop_percent'] / 30 * 100.0

user_thirty_popular_percentage.sort_values(by='thirty_pop_percent', ascending=False).plot()
```

```python
# Соотношение среднего времени прослушивания трека пользователем к
# общему среднему времени прослушивания трека
#- users_audition_quality['listening_coef']

# Делаем выборку из techno_data, включающую посчитанный выше столбец l_time_to_t_duration 
users_audition_quality = techno_data[['user_id', 'audition_id', 'track_id', 'l_time_to_t_duration']]

# Добавляем средний процент прослушивания треков mean_l_time_to_duration
users_audition_quality = users_audition_quality.join(track_quality, how='inner', on='track_id')

# calc вычисляем соотношение l_time_to_t_duration / mean_l_time_to_duration
users_audition_quality['listening_coef'] = users_audition_quality['l_time_to_t_duration'] / users_audition_quality['mean_l_time_to_duration']

# группируем вычисления по пользователем и вычисляем средние значения полученных выше соотношений
users_audition_quality = users_audition_quality[['user_id', 'listening_coef']].groupby('user_id').mean()

users_audition_quality.sort_values(by='listening_coef', ascending=False).plot()
```

Данные для скоринга готовы:
1. user_techno_to_all_tracks[techno_to_all_percent]
2. user_thirty_popular_percentage[thirty_pop_percent]
3. users_audition_quality[listening_coef]

Решил нормализовать их, прежде чем считать итоговый результат. Также добавил веса по-умолчанию, равные 1.0.

```python
# Вычисление итогового score

# Определяем веса
tech_percent_w = 1.0
thirty_pop_w = 1.0
aud_quality_w = 1.0

# Нормализуем данные
user_techno_to_all_tracks['techno_to_all_percent_n'] = user_techno_to_all_tracks['techno_to_all_percent'] / user_techno_to_all_tracks['techno_to_all_percent'].max()
user_thirty_popular_percentage['thirty_pop_percent_n'] = user_thirty_popular_percentage['thirty_pop_percent'] / user_thirty_popular_percentage['thirty_pop_percent'].max()
users_audition_quality['listening_coef_n'] = users_audition_quality['listening_coef'] / users_audition_quality['listening_coef'].max()

# вычисляем score
common_data = user_techno_to_all_tracks.join(user_thirty_popular_percentage, on='user_id', how='inner').join(users_audition_quality, on='user_id', how='inner')
common_data['score'] = common_data['techno_to_all_percent_n']*tech_percent_w + common_data['thirty_pop_percent_n']*thirty_pop_w + common_data['listening_coef_n']*aud_quality_w

common_data.sort_values(by='score', ascending=False)['score'].plot()
```

Далее считаем экономику - нам нужно обеспечить конверсию в покупку билета, такую чтобы затраты на привлечение клиента не превышали прибыль от продажи билета. 

Пусть стоимость билета - 700 рублей. С учетом всех затрат на проведение мероприятия определяется максимальная стоимость привлечения. Допустим, это 200р. 

Тогда, при стоимости рассылки одного письма в 10 рублей, конверсия в покупку должна быть не меньше 10/200*100% = 5%

Имея эти данные и наш Score, начинаем рассылку с тех, у кого score максимальный, и сразу начинаем собирать данные о конверсии из рассылки письма в покупку билета.
Продолжаем до тех пор, пока конверсия в покупку не опустится ниже порогового значения в 5%.
