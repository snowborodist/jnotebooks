Чтобы разобраться в ситуации с тарифами, построим модель:

```python
# Общие переменные
free_wait_min = 3            # количество бесплатных минут ожидания
cost_wait_min = 9            # стоимость минуты ожидания
distance = 16.5              # дистанция поездки

# UberX
u_min_cost = 129             # минимальная стоимость
u_included_min = 5           # включенные минуты
u_included_km = 4            # включенные километры
u_cost_min = 6               # тариф за минуту поездки
u_cost_km = 12               # тариф за километр поездки
u_wait_min = 3               # тариф за минуту ожидания
u_distance = distance        # дистанция поездки

# Яндекс.Go
y_low_demand_coef = 1.00     # коэффициент спроса (для расчета низкого спроса)
y_min_cost = 139             # минимальная стоимость
y_included_min = 6           # включенные минуты
y_included_km = 3            # включенные километры
y_cost_min = 10              # тариф за минуту поездки
y_cost_km = 8                # тариф за километр поездки
y_wait_min = 3               # тариф за минуту ожидания
y_distance = distance        # дистанция поездки
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (15, 10)
```

```python
ride_duration = np.arange(10, 40, 1)
avg_speed = distance / ride_duration * 60
```

```python
def uber_ride_cost(r_duration, total_distance = u_distance, wait_min = u_wait_min):
    return u_min_cost + (wait_min-free_wait_min)*cost_wait_min + (total_distance-u_included_km)*u_cost_km + (r_duration-u_included_min)*u_cost_min

def yandex_ride_cost(r_duration, total_distance = y_distance, wait_min = y_wait_min, low_demand_coef = y_low_demand_coef):
    return y_min_cost + (wait_min-free_wait_min)*cost_wait_min + (total_distance-y_included_km)*y_cost_km*low_demand_coef + (r_duration-y_included_min)*y_cost_min*low_demand_coef

def plot(x, x2, yandex_cost, uber_cost, xlabel):
    plt.plot(x, yandex_cost, x2, uber_cost)
    plt.xlabel(xlabel, color='gray')
    plt.ylabel('Ride Cost',color='gray')
    plt.legend(['Yandex.Go', 'UberX'])
    plt.grid(True)
```

```python
u_ride_cost = uber_ride_cost(ride_duration)
y_ride_cost = yandex_ride_cost(ride_duration)
```

```python
plot(avg_speed, avg_speed, y_ride_cost, u_ride_cost, 'Average Speed')
```

Модель готова.
По ней видно, что при низкой средней скорости поездки выгоднее оказывается тариф UberX, а если ехать быстро - Яндекс.Go.
При прочих равных условиях цены пересекаются в районе средней скорости равной около 64 км/ч. В городе такие цифры практически недостижимы, и почти всегда UberX будет выгоднее (за исключением случая, когда большая часть маршрута проходит по бессветофорной магистрали, а сами поездки совершаются при отсутствии машин, например ночью).

Но есть факторы, которые могут серьезно повлиять на ценовой расклад:
    1. у яндекса есть понижающий коэффициент (помимо повышающего, который по условию отсутствует)
    2. Есть фрод со стороны водителей
    3. Есть вариативность прокладывания маршрутов
    4. Есть ожидания (в начале поездки и в случае остановки по пути)
    5. Есть заезды в промежуточные пункты
    6. Есть другие факторы, которые я упустил =)
    
Рассмотрим эти варианты подробнее:

```python
#1 ПОНИЖАЮЩИЙ КОЭФФИЦИЕНТ
# При применении понижающего коэффициента 0.8 пересечение графиков наступает при средней скорости около 27 км/ч.
# Соответственно, при изменении коэффициента и при разной загрузке дорог (от этого колеблется средняя скорость)
# один из сервисов может становиться выгоднее другого, и наоборот.

ride_duration = np.arange(10, 100, 1)
avg_speed = distance / ride_duration * 60
u_ride_cost = uber_ride_cost(ride_duration)
y_ride_cost = yandex_ride_cost(ride_duration, low_demand_coef=0.8)
```

```python
plot(avg_speed, avg_speed, y_ride_cost, u_ride_cost, 'Average Speed')
```

```python
# 2,4,5 ФРОД СО СТОРОНЫ ТАКСИСТА / ВЫХОД ЗА 3 МИНУТЫ БЕСПЛАТНОГО ОЖИДАНИЯ ТАКСИ / ОЖИДАНИЕ В ПУТИ
# Если таксист попадется хитрый, и он сможет начать ожидание клиента раньше, чем положено,
# то клиент потом будет вынужден оплатить каждую лишнюю минуту простоя такси.
# На графике видно, что лишние 7 минут ожидания так же двигают пересечение кривых стоимости поездки
# в район значений около 32 км/ч.
# Тот же принцип справедлив, если один из наших коллег задержится в офисе и не сможет сразу выйти к такси.

ride_duration = np.arange(10, 100, 1)
avg_speed = distance / ride_duration * 60
u_ride_cost = uber_ride_cost(ride_duration, wait_min = 10)
y_ride_cost = yandex_ride_cost(ride_duration)
```

```python
plot(avg_speed, avg_speed, y_ride_cost, u_ride_cost, 'Average Speed')
```

```python
# 3 ВАРИАТИВНОСТЬ ПРОКЛАДЫВАНИЯ МАРШРУТОВ
# В зависимости от ситуации на дорогах, а также от наличия машин такси поблизости, 
# маршрут от работы до дома у двух коллег может быть проложен по-разному, и быть разной длины.
# Если у Яндекc.Go маршрут окажется длиннее всего на 4 км, чем у UberX, пересечение графиков
# стоимости передвинется к значению около 28 км/ч.

y_distance = 20.5
ride_duration = np.arange(10, 100, 1)
avg_speed = distance / ride_duration * 60
avg_speed2 = y_distance / ride_duration * 60
u_ride_cost = uber_ride_cost(ride_duration, wait_min = 10)
y_ride_cost = yandex_ride_cost(ride_duration, total_distance = y_distance)
```

```python
plot(avg_speed, avg_speed2, y_ride_cost, u_ride_cost, 'Average Speed')
```

Вывод:  
Самые разные ситуации могут существенно влиять на место пересечения графиков цен поездок, и тем самым то один, то другой тариф может оказываться выгоднее.
